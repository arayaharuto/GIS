```python
fig = px.bar(gdf.head(10),
            y='JINKO',
            x='S_NAME',
            template='seaborn' # change this to see other styles
            )

fig.show()
```

```python
# population map
ax = gdf.head(100).plot(figsize=(10,10),
         column='JINKO',
         legend=True,
         cmap='autumn_r',
         edgecolor='black',
         linewidth=1,
         alpha=0.8)

# hide axes
ax.axis('off')

# title
ax.set_title('市川市の人口の多い街トップ100',fontsize=24);

# add basemap
cx.add_basemap(ax,source=cx.providers.Stamen.TonerLite)
```

世帯数や人口密度が高かったり低かったりするところを実際に調べてみると理由がちゃんとあって考察できるのが面白かった。
