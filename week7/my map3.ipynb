```python
import geopandas as gpd
import pandas as pd

niigata_ku = gpd.read_file('N03-23_15_230101.geojson')

niigata_ku.plot()

niigata_ku.head()

niigata_ku.info()

niigata_pop = pd.read_csv('niigata.csv')

niigata_pop.head()

niigata_pop.info()

niigata_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
nigata_ku

niigata_ku['KEY_CODE'] = niigata_ku['KEY_CODE'].astype(int)

niigata_ku.info()

niigata_ku_pop = niigata_ku.merge(niigata_pop,on='KEY_CODE')

niigata_ku_pop

niigata_ku_pop.info()

niigata_ku_pop['人口総数'] = niigata_ku_pop['人口総数'].astype(int)

niigata_ku_pop.info()

niigata_ku_pop.plot(column='人口総数',legend=True)


```
